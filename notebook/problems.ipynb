# Hypothesis testing problems

## Exercise 1

You are a nutritionist researching two different types of diets to see if there is a significant difference in weight loss after one month. You choose two random groups of people; one group follows the first diet, and the other follows the second. At the end of the month, you record the weight loss (in kg) for each person:

| Diet 1 | Diet 2 |
|:-------|:-------|
| 2.0 | 3.0 |
| 2.5 | 3.2 |
| 3.0 | 3.1 |
| 2.8 | 2.9 |
| 2.3 | 2.8 |
| 2.7 | 3.0 |
| 2.5 | 3.2 |

With these data, it seeks to answer the following question: Is there a significant difference in average weight loss between people who followed the first diet and those who followed the second diet?

To draw conclusions, follow the points below:

- State the hypothesis: null and alternative hypothesis.
- Perform the test to test the hypothesis. You can use a Student's t-test.
- Analyze the conclusions.

## Vamos a formular las hipótesis:
1- Hipótesis nula: no existen diferencias significativas entre las dietas 
2- Hipótesis alternativa: existen diferencia significativas entre las dietas

```python
# My code
import scipy.stats as stats
import numpy as np
from statsmodels.stats.multicomp import pairwise_tukeyhsd

dieta_1 = [2.0, 2.5, 3.0, 2.8, 2.3, 2.7, 2.5]
dieta_2 = [3.0, 3.2, 3.1, 2.9, 2.8, 3.0, 3.2]

# Aplicamos Student's t-test
t_value, p_value = stats.ttest_ind(dieta_1, dieta_2)

#Llegamos a la conclusion
print(f"t-value: {t_value}")
print(f"p-value: {p_value}")

if p_value < 0.05:
    if t_value < 0:
        conclusion = ("Los resultados indican que hay diferencias significativas entre los dos grupos estudiados "
                      "ya que el p-valor es menor a 0.05. Además, la media del segundo grupo es mayor que la del primero, "
                      "lo cual se evidencia por el valor negativo del t.")
    else:
        conclusion = ("Los resultados indican que hay diferencias significativas entre los dos grupos estudiados "
                      "ya que el p-valor es menor a 0.05. Además, la media del primer grupo es mayor que la del segundo, "
                      "lo cual se evidencia por el valor positivo del t.")
else:
    conclusion = ("Los resultados indican que no hay diferencias significativas entre los dos grupos estudiados "
                  "ya que el p-valor es mayor o igual a 0.05.")

print(conclusion)
```

## ANOVA

**ANOVA** (*Analysis of Variance*) is a statistical technique used to compare the measures of two or more groups. The idea behind ANOVA is to decompose the total variability in the data into two components: between-group variability and within-group variability:

- **Between-group variability**: This variability refers to the differences between the group means. If this variability is considerably larger than the within-group variability, it could be an indication that at least one of the group means is different.
- **Within-group variability**: This variability refers to the dispersion of the data within each group. If all groups have similar variability, then any noticeable difference in group means could be considered significant.

Hypotheses in ANOVA typically include:

- **Null hypothesis** ($H₀$): The means of all groups are equal.
- **Alternative hypothesis** ($H₁$): At least one of the group means is different.

If the ANOVA test result is significant (e.g., a p-value less than a threshold such as 0.05), this suggests that at least one group mean is different.

## Exercise 2

A farmer decides to test three different types of fertilizers to determine if one is superior in terms of corn production. The farmer plants corn on 15 identical plots and uses all three fertilizers (5 plots for each type). At the end of the season, he measures the corn yield (in kg) of each plot, with the following result:

| Fertilizer 1 | Fertilizer 2 | Fertilizer 3 |
|:-------------|:-------------|:-------------|
| 20 | 22 | 24 |
| 21 | 21 | 23 |
| 20 | 23 | 22 |
| 19 | 22 | 23 |
| 20 | 21 | 24 |

With this data, he seeks to answer the following question: Is there a significant difference in average corn yield between the three types of fertilizers?

To help you, follow the points below:

- State the hypothesis: null and alternative hypothesis.
- Perform the ANOVA test.
- Analyze the conclusions.
- If one fertilizer is better than another, how can we know it?

## Vamos a formular las hipótesis:
1- Hipótesis nula: no existen diferencias significativas en el rendimiento de los fertilizantes 
2- Hipótesis alternativa: existen diferencia significativas en el rendimiento de los fertilizantes 

```python
# My code

fertilizante_1 = [20, 21, 20, 19, 20]
fertilizante_2 = [22, 21, 23, 22, 21]
fertilizante_3 = [24, 23, 22, 23, 24]

#Procedemo con el ANOVA test
f_value, p_value = stats.f_oneway(fertilizante_1, fertilizante_2, fertilizante_3)

#Concluimos en:
print(f"f-value: {f_value}")
print(f"p-value: {p_value}")

if p_value < 0.05:
    conclusion = ("Los resultados del ANOVA indican que existen diferencias significativas entre el rendimiento de los "
                  "diferentes fertilizantes, ya que el p-valor es menor a 0.05. Esto sugiere que al menos uno de los "
                  "fertilizantes tiene un efecto diferente en el rendimiento comparado con los otros.")
else:
    conclusion = ("Los resultados del ANOVA indican que no existen diferencias significativas entre el rendimiento de los "
                  "diferentes fertilizantes, ya que el p-valor es mayor o igual a 0.05. Esto sugiere que los fertilizantes "
                  "tienen efectos similares en el rendimiento.")

print(conclusion)
```

```python
#Combinamos los datos en un solo array
data = np.concatenate([fertilizante_1, fertilizante_2, fertilizante_3])
labels = ["F1"] * len(fertilizante_1) + ["F2"] * len(fertilizante_2) + ["F3"] * len(fertilizante_3)

#VAmos a realizar el test de Tukey HSD
result = pairwise_tukeyhsd(data, labels, alpha=0.05)
print(result)

#Interpretamos  los resultados
interpretacion = []

for comparison in result._results_table.data[1:]:
    group1, group2, meandiff, p_adj, lower, upper, reject = comparison
    conclusion = f"Comparación {group1} vs {group2}: "
    if reject:
        conclusion += f"existe una diferencia significativa (p-adj = {p_adj:.4f}, intervalo de confianza = [{lower:.4f}, {upper:.4f}])."
    else:
        conclusion += f"no existe una diferencia significativa (p-adj = {p_adj:.4f}, intervalo de confianza = [{lower:.4f}, {upper:.4f}])."
    interpretacion.append(conclusion)

print("\nInterpretación de los resultados:")
for conclusion in interpretacion:
    print(conclusion)

# Determinación del mejor y peor fertilizante basado en las medias
medias = {
    "F1": np.mean(fertilizante_1),
    "F2": np.mean(fertilizante_2),
    "F3": np.mean(fertilizante_3)
}

mejor_fertilizante = max(medias, key=medias.get)
peor_fertilizante = min(medias, key=medias.get)

print(f"\nEl fertilizante con mejor rendimiento es {mejor_fertilizante} con una media de {medias[mejor_fertilizante]:.2f}.")
print(f"El fertilizante con peor rendimiento es {peor_fertilizante} con una media de {medias[peor_fertilizante]:.2f}.")
```
